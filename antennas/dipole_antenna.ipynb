## Dipole

### Infinitesimal dipole
In this case we take the lenght of a linear wire very small $l << \lambda$ and very thin $a << \lambda$. The first approach is to take constant value of the current accross the wire. This is one of the few problems that can be solved with pure algebra and the process is as follow:

1. Calculate the vector potential

$$\vec{A(x)} = \frac{\mu}{4\pi} \int I(\vec{x'}) \frac{e^{-jkR}}{R} dl' = \hat{z}\frac{\mu I_0 l}{r\pi r} e^{-jkr}$$

Transforming the potential to spherical we obtain:
$$A_r = \frac{\mu I_0 l}{r\pi r} e^{-jkr} \cos(\theta) $$
$$A_{\theta} = -\frac{\mu I_0 l}{r\pi r} e^{-jkr} \sin(\theta)$$
$$A_{\phi} = 0 $$

2. Compute $\vec{H}$
$$H_r = H_{\theta} = 0 $$
$$H_{\phi} = j \frac{k I_0 l \sin(\theta)}{4\pi r} \left[1+ \frac{1}{jkr} \right] e^{-jkr} $$


3. Compute $E$
$$ E_r = \eta \frac{I_0 l \cos(\theta)}{2\pi r^2} \left[1+\frac{1}{jkr} \right] e^{-jkr}$$
$$ E_{\theta} = -j \eta \frac{k I_0 l \sin(\theta)}{2\pi r} \left[1+\frac{1}{jkr} - \frac{1}{k^2 r^2} \right] e^{-jkr} $$
$$ E_{\phi} = 0$$

5. Calculate the power density using the pointing vector
$$W_r = \frac{\eta}{8} \left| \frac{I_0 l}{\lambda} \right| \frac{\sin(\theta)^2}{r^2} \left[1- j\frac{1}{k^3r^3} \right]  $$
$$W_{\theta} = j\eta \frac{k|I_0 l|\cos(\theta)\sin(\theta) }{16\pi^2 r^3 } \left[1+\frac{1}{k^2r^2} \right] $$


Of all this math one important thing must be noted.. the power in the $\theta$ axis is purely reactive, also in the $\vec{r}$ we got an imaginiary component that is very important at small values of $kr$. This is the perfect example of the radiation zones (reactive, near field, far field).

Usually we just care about the far field where all the $1/kr$ terms vanish.

### Finite sized dipole

To solve this problem we suppose that the finite dipole is build by the sum of infinitesimal dipoles, in far field:
$$ dE_{\theta} \sim j\eta \frac{k I(\vec{x'})e^{-jkR} }{r\pi R}  \sin(\theta) dz'$$
$$ dH_{\phi} \sim j \frac{kI(\vec{x'})e^{-jkR} }{4\pi R} \sin(\theta) dz'$$
$$ dE_{r} = dE_{\phi} = dH_{r} = dH_{\theta} \sim 0 $$


Then the $E_{\theta}$ becames the following expression, where we can identify two expressions. The one between parenthesis that is called the space factor and the one outside the parenthesis that is called element factor.
$$ E_{\theta} = \int_{-l/2}^{l/2} dE_{\theta} =  j\eta \frac{ke^{-jkr}}{4\pi r} \sin(\theta) \left[ \int_{-l/2}^{l/2} I e^{jkz'\cos(\theta)}  dz' \right]$$

### Half-wave dipole
One of the most common antennas, its radiation resistance is $73 \Omega$. Another neat property is that around this length value the reactance also goes to zero.
$$ E_{\theta} \sim j \eta \frac{I_0e^{jkr}}{2\pi r} \left[ \frac{\cos(\pi/2 \cos(\theta))}{\sin(theta)} \right]$$
$$ H_{\phi} \sim j \frac{I_0e^{jkr}}{2\pi r} \left[ \frac{\cos(\pi/2 \cos(\theta))}{\sin(theta)} \right]$$

$$W =  \eta \frac{I_0^2}{8\pi^2r^2} \left[ \frac{\cos(\pi/2 \cos(\theta))}{\sin(theta)}  \right]^2 \approx \eta \frac{I_0^2}{8\pi^2r^2} \sin(\theta)^3 $$

Then $U \sim \sin^3(\theta)$

### Rules of thumbs
1. To consider a dipole infinitesimal $l<\lambda/50$, a short dipole is when $l<\lambda/10$
2. It is a vertical polarized antenna.
3. Short dipoles had a very low real impedance and high imaginary impedance.
4. Maximum gain is at $\theta = \pi/2$ that is the plane $x,y$.
5. For a finite size dipole when the lenght approach $\lambda$ its directivity increase but also its input impedance goes super high..
6. For $l<<\lambda$ you get a HPBW of $\approx 90^\circ$, at $l=\lambda$ you have HPWB of $47.8^{\circ}$
7. When having a lenght higher than $\lambda$ then secondary lobes appears.
8. The impedance goes to infinity when the lenght approach $\lambda$ and its multiples.
9. The gap at the input terminals of the dipole play a major role in the actual input impedance (you have to imagine that is the parameter that allows you to play with the current distribution on the dipole). For example you can have one of the wires longer than the other one to avoid having $\infty$ impedance at feed. The other parameter that you can tune is the radius of the wire.
10. Depending on the radius of the wire the resonance can be slightly shifted .
11. Making the dipole more fat (increasing the radius) you will increase the bandwidth of it.. but you start to move the resonant frequency down.
12. The standard bandwidth of these antennas is around $3-10%$.
