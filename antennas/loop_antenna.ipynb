## Loop antennas

### Small circular loop

We take a small wire forming a loop in the plane x-y at z=0 with a radius $a$. Again we take constant current $I_{\phi}=I_0$ (this is only valid for the small antenna).

1. Calculate the vector potential
First he moves everything to spherical coordinates ad write the current distribution.. After some algebra we got:
$$ I_{e} = I_{\phi} \sin(\theta)\sin(\phi'-\phi)\hat{r} + I_{\phi}\cos(\theta)\sin(\phi'-\phi)\hat{\theta}+I_{\phi}\cos(\phi'-\phi)\hat{\phi}$$

Replacing this in the equation for $A$ and writing $R=\sqrt{r^2+a^2-2ar\sin(\theta)\cos(\phi'-\phi)}$ we got an awfull integral:

$$A_{\phi} = \frac{a\mu}{4\pi}\int_0^{2\phi} I_{\phi} \cos(\phi'-\phi) \frac{e^{-jk\sqrt{r^2+a^2-2ar\sin(\theta)\cos(\phi'-\phi)}}}{\sqrt{r^2+a^2-2ar\sin(\theta)\cos(\phi'-\phi)}} d\phi' $$

Like $I_{phi}$ is constant you can take it out of the integral..but still need to solve the awfull one. For that he uses Taylor series and solve for the first two terms to obtain the following value:

$$A_{\phi} \approx \frac{a^2\mu I_0}{4\pi}e^{-jkr} \left(\frac{jk}{r}+\frac{1}{r^2} \right)\sin(\theta) $$ 
$$A_{r}= A_{\theta} \approx 0 $$

2. Compute $\vec{H}$
$$H_{r} = j\frac{ka^2 I_0 \cos(\theta)}{2r^2}\left[1+\frac{1}{jkr} \right]e^{-jkr} $$
$$H_{\theta} = -\frac{(ka)^2I_{0}\sin(\theta)}{4r}\left[1+\frac{1}{jkr}-\frac{1}{(kr)^2}\right]e^{-jkr} $$
$$H_{\phi}=0 $$

3. Compute $\vec{E}$
$$E_{r} = E_{\theta} = 0 $$
$$E_{\phi} = \eta \frac{(ka)^2I_0\sin(\theta)}{4r} \left[1+\frac{1}{jkr} \right]e^{-jkr} $$

One nice thing to note is that this results are the same ones that you would have if you use a magnetic dipole with magnetic moment $I_m$ that satisfy $I_m\cdot l = jS\omega \mu I_0$ where $S=\pi a^2$ is the area of the loop.


4. Calculate the power density
Again we have two marked zones, the first one where $kr<<1$ and the power is mostly reactive and the second one where $kr>>1$ where the power is mostly real.

$$ W_r = \eta \frac{(ka)^4}{32} I_0^2 \frac{\sin(\theta)^2}{r^2} \left[ 1+j\frac{1}{(kr)^3} \right]$$

One key difference between the loop antenna and the dipole is that the reactive part is now purely inductive (positive sign) and in the dipole its capacitive.


The resistance of the small loop is given by the nex equation where N is the number of turns
$$ R_r = \eta \left( \frac{2\pi}{3} \right) \left(\frac{kS}{\lambda} \right)^2 N^2 \approx 31.171 N^2 \frac{S^2}{\lambda} $$

### Circular loop of constant current 
Now consider a loop that is not small, in this case he cant do the taylor expansion.. and need to deal with the awfull integral.. The solution involve some Bessel functions (that are typical for cyclindrical problems).

$$A_{\phi} = j\frac{a\mu I_0e^{-jkr}}{2r}J_1(ka\sin(\theta)) $$

$$E_{r} = E_{\theta} = H_r = H_{\phi} = 0 $$
$$E_{\phi} \approx \frac{ak\eta I_0e^{-jkr}}{2r} J_1(ka \sin(\theta)) $$
$$H_{\theta} \approx -\frac{ak I_0 e^{-jkr}}{2r}J_1(ka\sin(\theta)) $$


The radiation intensity is given then by
$$ W_{avg} = \hat{r} \frac{\left(a\omega \mu I_0 \right)^2}{8\eta r^2}J_1^2(ka\sin(\theta)) $$

## Rules of thumbs
1. You can increase the resistance of the loop antenna making several turns (should be quadratic with the number of turns). Also another method is to use ferrite.
2. The loss resistance of a loop antenna is much larger than the radiation resistance, so the efficiency of this antenna is low. To avoid that multiple turns can be made, but this will change the current distribution (need to be simulated..)
3. Again the physical size of the loop is smaller than its electrical size....
4. The loop antenna is used typically in receiving mode bcs is omni. Also its used as probe to measure magnetic flux.
5. The small antenna approx works for circunference less than $0.2\lambda$ or radius $a<0.032\lambda$.
6. For constant current assumption, when increasing the radius $a$ the main lobe increases. Passed the $0.5\lambda$ you start get sidelobes. (The point for start getting sidelobes is at $a=0.61\lambda$.
7. Radiation resistance increase with the radius of the wire, also the directivity (but you got the sidelobes).
8. When the circunference approach $\lambda$ the uniform current approx stop working and the beam pattern changes and points to $\theta=0^\circ,180^\circ$
9. In non-uniform current approach, there is a huge values on the impedance for $C/\lambda = 0.5$ and the first resonant value (when the reactance is 0) its at roughly $C\lambda = 1$
10. Some times, to avoid effects that feed could introduce into the loop antenna a balun is used.
11. The rule of thumb is to start designing with a circunference of $1\lambda$.. you could play with the actual geometry to minimize the area used, but could be detrimental too....
